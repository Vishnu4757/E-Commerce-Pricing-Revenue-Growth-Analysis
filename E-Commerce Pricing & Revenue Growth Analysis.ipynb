## **Understand Each Column :**

- **Order_id :-**

            - A unique identifier assigned to each transaction in the e-commerce platform.
            - This is the key target column for risk analysis.
  
- **Order_date :-**
    
            - Represents the date when a customer placed the order on the platform.
            - This column is useful for analyzing monthly, yearly, and seasonal sales trends.
  
- **State :-**

            - The amount of money involved in the current transaction.
            - It helps analyze regional sales performance and geographic revenue distribution..

  
- **Zone :-**

            - Represents the broader geographical region of India such as North, South, East, or West.
            - This helps identify which regions contribute the most to overall sales and revenue.

- **category :-** 
                
            - Refers to the product category such as Electronics, Fashion, Home Appliances, etc.
            - It allows businesses to analyze which product categories generate the highest sales.

 
- **Brand_type :-** 
    
            - Indicates whether the product belongs to a Premium brand or a Budget brand.
            - This helps evaluate how different pricing segments influence customer purchasing behavior.

  
- **Customer_gender :-** 
    
            - Represents the gender of the customer who made the purchase (Male or Female).
            - This column helps analyze gender-based purchasing patterns and revenue contribution.

  
- **Customer_age :-** 
                
            - Shows the age of the customer placing the order.
            - It helps identify which age groups are the most active buyers on the platform.

  
- **Base_price :-** 
    
            - The original price of the product before any discounts are applied.
            - It helps analyze the initial pricing strategy of products.

  
- **Discount_percent :-**
    
            - Represents the percentage of discount offered on a product.
            - This column helps analyze how discounts influence customer buying behavior and sales volume. 

  
- **Final_price :-**
                                                           
            - The price that the customer actually pays after applying the discount.
            - It reflects the effective selling price of the product.

  
- **Units_sold :-**

            - Indicates the number of units sold for a particular product in a transaction.
            - It helps measure product demand and sales volume.


- **Revenue :-** 
    
            - Represents the total revenue generated from the sale of the product.
            - It is calculated based on the final price and number of units sold.

  
- **Sales_event :-**
    
            - Indicates whether the purchase occurred during a promotional event such as festival sales or special offers.
            - This helps analyze the impact of marketing campaigns and seasonal sales on revenue. 

  
- **Competition_intensity :-**
                                                           
            - Represents the level of competition in the market for a particular product category.
            - It helps understand how competitive pressure influences pricing and sales performance.

  
- **Inventory_pressure :-**

            - Indicates the level of inventory demand or stock pressure for products.
            - This helps businesses understand supply-demand balance and inventory management challenges.

## **Handling Outliers :**

```python
numeric_cols=data_clean.select_dtypes(include=['number']).columns
numeric_cols

print("\n === Outliers Count(IQR method) ===")

for col in numeric_cols:
    Q1= data_clean[col].quantile(0.25)
    Q3= data_clean[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    outliers = data_clean[(data_clean[col] < lower) | (data_clean[col] > upper)]
    print(f"{col} : {len(outliers)} outliers")
```

* As we see they are some outliers are present in dataset.
* From the above dataset:
   * The Revenue has highest outliers with 2173 outliers
   * The Customers_Age and Discount_Percent has no outliers
   * And Remaining all columns has outliers 

```python
Q1 = data_clean['Base_Price'].quantile(0.25)
Q3 = data_clean['Base_Price'].quantile(0.75)
IQR = Q3 - Q1

lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR

data_clean.loc[data_clean['Base_Price'] > upper_limit, 'Base_Price'] = upper_limit
data_clean.loc[data_clean['Base_Price'] < lower_limit, 'Base_Price'] = lower_limit
```

```python
Q1 = data_clean['Final_Price'].quantile(0.25)
Q3 = data_clean['Final_Price'].quantile(0.75)
IQR = Q3 - Q1

lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR

data_clean.loc[data_clean['Final_Price'] > upper_limit, 'Final_Price'] = upper_limit
data_clean.loc[data_clean['Final_Price'] < lower_limit, 'Final_Price'] = lower_limit
```

```python
Q1 = data_clean['Units_Sold'].quantile(0.25)
Q3 = data_clean['Units_Sold'].quantile(0.75)
IQR = Q3 - Q1

lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR

data_clean.loc[data_clean['Units_Sold'] > upper_limit, 'Units_Sold'] = upper_limit
data_clean.loc[data_clean['Units_Sold'] < lower_limit, 'Units_Sold'] = lower_limit
```

```python
Q1 = data_clean['Revenue'].quantile(0.25)
Q3 = data_clean['Revenue'].quantile(0.75)
IQR = Q3 - Q1

lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR

data_clean.loc[data_clean['Revenue'] > upper_limit, 'Revenue'] = upper_limit
data_clean.loc[data_clean['Revenue'] < lower_limit, 'Revenue'] = lower_limit
```
